```julia
function euler_step(x0, y0, f, h)
    return y0 .+ h.*f(x0, y0)
end
```

```julia
xs = collect(0:0.01:3)
ys = exp.(-xs)
plot(xs, ys)
```

$y = \exp(-x)$ satisfies 
$$
\frac{dy}{dx} = -y
$$
with $y(0) = 1$.

```julia
function f_exp(x, y)
    return -y
end
```

```julia
function euler_solve(x0, y0, f, x1, h)
    x = x0
    y = y0
    
    xs = [x0]
    ys = [y0]
    while x < x1
        y = euler_step(x, y, f, h)
        x = x+h
        push!(xs, x)
        push!(ys, y)
    end
    
    xs, ys
end
```

```julia
xs, ys = euler_solve(0.0, [1.0], f_exp, 10.0, 2.1)
```

```julia
plot(xs, [y[1] for y in ys])
plot(xs, exp.(-xs))
```
